```python

import seaborn as sns
import matplotlib.pyplot as plt
```

**Set a Seaborn theme for better aesthetics**


```python
sns.set_theme(style="whitegrid")
```

**Plot 1: Weather Condition Frequency**

```python
plt.figure(figsize=(14, 6))
sns.countplot(y='Weather_Condition', data=data, order=data['Weather_Condition'].value_counts().index)
plt.title("Frequency of Different Weather Conditions")
plt.xlabel("Count")
plt.ylabel("Weather Condition")
plt.show()
```

**Plot 2: Temperature Distribution by Weather Condition (Top 10)**

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Weather_Condition', y='Temp_C', data=data, order=data['Weather_Condition'].value_counts().index[:10])
plt.title("Temperature Distribution by Weather Condition (Top 10)")
plt.xlabel("Weather Condition")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.show()
```

**Plot 3: Scatter Plot of Humidity vs. Temperature**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temp_C', y='Rel Hum_%', data=data, hue='Weather_Condition', palette='cool', legend=False)
plt.title("Humidity vs. Temperature Scatter Plot")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Relative Humidity (%)")
plt.show()
```

**Plot 4: Visibility under Different Weather Conditions (Top 10)**

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Weather_Condition', y='Visibility_km', data=data, order=data['Weather_Condition'].value_counts().index[:10])
plt.title("Visibility Distribution by Weather Condition (Top 10)")
plt.xlabel("Weather Condition")
plt.ylabel("Visibility (km)")
plt.xticks(rotation=45)
plt.show()
```

**Plot 5: Temperature Trends Over Time**

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x='Date&Time', y='Temp_C', data=data)
plt.title("Temperature Trends Over Time")
plt.xlabel("Date & Time")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.show()
```

**Plot 6: Wind Speed Distribution**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Wind Speed_km/h'], bins=20, kde=True)
plt.title("Distribution of Wind Speed")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Frequency")
plt.show()
```

**Plot 7: Humidity vs. Visibility**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rel Hum_%', y='Visibility_km', data=data, hue='Weather_Condition', palette='viridis', legend=False)
plt.title("Humidity vs. Visibility")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Visibility (km)")
plt.show()
```

**Plot 8: Correlation Heatmap**

```python
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Weather Variables")
plt.show()
```

**Plot 9: Pair Plot of Key Features**

```python
sns.pairplot(data[['Temp_C', 'Rel Hum_%', 'Visibility_km', 'Wind Speed_km/h']], diag_kind='kde')
plt.suptitle("Pair Plot of Key Weather Features", y=1.02)
plt.show()
```
